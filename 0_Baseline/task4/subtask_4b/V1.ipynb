```python
import json
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize
import nltk
from tqdm import tqdm

# Download tokenizer (only needed once)
nltk.download("punkt")

# --- Load Data ---
with open("task4b_train.json") as f:
    train_data = json.load(f)

with open("task4b_papers.json") as f:
    papers_data = json.load(f)

# --- Prepare Documents ---
paper_ids = []
doc_texts = []

for paper in papers_data:
    paper_ids.append(paper["cord_uid"])
    text = paper.get("title", "") + " " + paper.get("abstract", "")
    doc_texts.append(text.lower())

tokenized_corpus = [word_tokenize(doc) for doc in doc_texts]

# --- Build BM25 Index ---
bm25 = BM25Okapi(tokenized_corpus)

# --- Retrieve Top 5 Documents per Tweet ---
results = []

for item in tqdm(train_data):
    tweet_id = item["id"]
    tweet_text = item["tweet"].lower()
    tokenized_query = word_tokenize(tweet_text)

    scores = bm25.get_scores(tokenized_query)
    top_indices = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)[:5]
    top_doc_ids = [paper_ids[i] for i in top_indices]

    results.append({
        "id": tweet_id,
        "doc_ids": top_doc_ids
    })

# --- Save Results to Submission Format ---
with open("submission_bm25.json", "w") as f:
    json.dump(results, f, indent=2)

print("âœ… Submission file saved as submission_bm25.json")
```
